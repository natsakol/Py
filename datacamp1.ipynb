```python
import quandl
aapl = quandl.get("WIKI/AAPL", start_date="2015-06-19", end_date="2018-07-19")
aapl.describe()
aapl.tail()
```

```python
import pandas as pd
print(aapl.loc[pd.Timestamp('2017-01-01'):pd.Timestamp('2017-02-28')].tail())
```

```python
sample = aapl.sample(5)
print(sample)
```

```python
monthly_aapl = aapl.resample('M').mean()
print(monthly_aapl)
```
